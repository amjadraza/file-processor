```python
import pandas as pd
import json
import os
data_path = '../test_data/'
```

```python
import re
regex = r'^(-?(?:[1-9][0-9]*)?[0-9]{4})-(1[0-2]|0[1-9])-(3[01]|0[1-9]|[12][0-9])T(2[0-3]|[01][0-9]):([0-5][0-9]):([0-5][0-9]Z)?$'
match_iso8601 = re.compile(regex).match
def validate_iso8601_utc(str_val):
    try:            
        if match_iso8601( str_val ) is not None:
            return True
    except:
        pass
    return False
```

```python
def filter_json_date(file_name, start_datetime, end_datetime):
    data_path = '../test_data/'
    
    if validate_iso8601_utc(start_datetime) & validate_iso8601_utc(end_datetime):
        df = pd.read_csv(data_path + file_name, sep=" ",header=None)
        df.columns = ['eventTime', 'email', 'sessionId']
        
        return df[df['eventTime'].between(start_datetime, end_datetime)].to_json(orient='records', lines=False)
    else:
        print('date time format is not correct')
        
    
```

```python
file_name = 'sample3.txt'
start_datetime = '2000-01-01T17:25:49Z'
end_datetime = '2000-01-06T06:27:36Z'
filter_json_date(file_name, start_datetime, end_datetime)
```
